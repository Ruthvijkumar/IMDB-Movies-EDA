### The correlation matrix for numerical features in the dataset?

```python
import seaborn as sns

# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = df[numeric_columns]

# Correlation matrix for numerical features
correlation_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix for Numerical Features')
plt.show()
```

### Relation of numerical features in the dataset to each other

```python
import seaborn as sns

# Select only numeric columns for pair plot
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = df[numeric_columns]

# Pair plot for numerical features
sns.pairplot(numeric_data)
plt.suptitle('Pair Plot for Numerical Features', y=1.02)
plt.show()
```
